## WorldRemit Cohort Python 

To improve my pandas knowledge, I will try to recreate the same tables I created in SQL using Panda dataframes, Ill then run the same Seaborn script to create the visualisations

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import matplotlib.colors as mcolors
import numpy as np
```

```python
# Lets look at the raw data file and import to dataframe
wr = pd.read_csv('wr-data.csv')
wr.head(5)
```

```python
#We dont need the country code columns
wr.drop(columns=['send_country_code', 'receive_country_code'], inplace=True, axis=1)
```

```python
wr.head(5)
```

That's bettter!

```python
#lets find the first month the customer made a payment
month_customer = wr[['month','customer_id']]
```

```python
first_month_per_customer = month_customer.groupby(['customer_id']).min().reset_index()
```

```python
#Sense check the dataframe
first_month_per_customer.info()

# Number of rows is the same as the number of customers so thats good!
# lets rename the column 
first_month_per_customer.columns = ['customer_id', 'first_month']
first_month_per_customer.head(1)
```

```python
# Now lets join first_month back onto the wr dataframe
wr_with_first_month = pd.merge(wr, first_month_per_customer, how='inner', on='customer_id')
```

```python
#Also check if the customer is type 1 or type 2 only
customer_type_df = month_customer = wr[['customer_id', 'transaction_type', 'transaction_id']]
```

```python
customer_type_df.head()
customer_type_df_pivot = customer_type_df.pivot_table(index='customer_id', columns='transaction_type', values='transaction_id', aggfunc='count')
customer_type_df_pivot = customer_type_df_pivot.fillna(0)
customer_type_df_pivot
```

```python
wr_with_first_month['Period'] = wr_with_first_month['month'] - wr_with_first_month['first_month']
sorted_wr = wr_with_first_month.sort_values(by=['customer_id','month'])
sorted_wr.head(50)
```

```python
cohort = pd.merge(wr_with_first_month, customer_type_df_pivot, how='inner', on='customer_id')
cohort.columns = ['Month', 'Transaction_ID', 'Transaction_Type', 'Customer_ID', 'Revenue_GBP', 'First_Month', 'Period', 'Type 1', 'Type 2']
```

```python
type_1_customers = cohort.loc[(cohort['Type 1'] > 0)&(cohort['Type 2'] == 0) ].reset_index()
type_2_customers = cohort.loc[(cohort['Type 2'] > 0)&(cohort['Type 1'] == 0) ].reset_index()
```

```python
type_1_cohort = type_1_customers.groupby(['Month', 'First_Month', 'Period']).agg(
     count_cust = ('Customer_ID','nunique'),
     sum_rev     = ('Revenue_GBP','sum'),
     count_tx     = ('Transaction_ID','count'))
type_2_cohort = type_2_customers.groupby(['Month', 'First_Month', 'Period']).agg(
     count_cust = ('Customer_ID','nunique'),
     sum_rev     = ('Revenue_GBP','sum'),
     count_tx     = ('Transaction_ID','count'))
```

```python
type_1_cohort
```

```python
type_2_cohort
```

```python
cohort_pivot_type_1 = type_1_cohort.pivot_table(index = 'First_Month',
                                     columns = 'Period',
                                     values = 'count_cust')
cohort_pivot_type_2 = type_2_cohort.pivot_table(index = 'First_Month',
                                     columns = 'Period',
                                     values = 'count_cust')
```

```python
#Create the data frames for each transaction type

cohort_size_1 = cohort_pivot_type_1.iloc[:,0] # take the first column of of cohort df
retention_matrix_1 = cohort_pivot_type_1.divide(cohort_size_1, axis = 0) #calculate the retention percentage for each cohort

cohort_size_2 = cohort_pivot_type_2.iloc[:,0] # take the first column of of cohort df
retention_matrix_2 = cohort_pivot_type_2.divide(cohort_size_2, axis = 0) #calculate the retention percentage for each cohort

```

```python
#Ideally I would make this a function to reuse but time is money!

with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix_1, 
                mask=retention_matrix_1.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention, Type 1 Transactions', fontsize=16)
    ax[1].set(xlabel='# of months',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohort_size_1).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()

    
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix_2, 
                mask=retention_matrix_2.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention, Type 2 Transactions', fontsize=16)
    ax[1].set(xlabel='# of months',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohort_size_2).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()
```
