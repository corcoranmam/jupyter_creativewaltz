### How many customers joined each month and stayed?

Using the 'cohort_df's we imported from SQL we can pivot the data to look at the retention for each cohort. Here I've
used the month they made their first payment as the cohort since it is one of the common behaviours/attributes we have on the customers.

```python
cohort_pivot_type_1 = cohort_df_type_1.pivot_table(index = 'Cohort (Join Month)',
                                     columns = 'Period',
                                     values = 'total_customers')
cohort_pivot_type_2 = cohort_df_type_2.pivot_table(index = 'Cohort (Join Month)',
                                     columns = 'Period',
                                     values = 'total_customers')

cohort_pivot_type_1
```

```python
cohort_pivot_type_2
```

Here we can see how each cohort of customers returns each month to make a payment. Its not very pretty though is it? We can fix that

Now we can create the nice graphic. We create two new dataframes. One will hold the the number of customers who made a first payment in the month.
The second is the percentage increase/decrease of returning customers each month after

I have in experience of creating nice visualizations so altered the code from this site:
https://towardsdatascience.com/a-step-by-step-introduction-to-cohort-analysis-in-python-a2cbbd8460ea
to create my tables

```python
#Create the data frames for each transaction type

cohort_size_1 = cohort_pivot_type_1.iloc[:,0] # take the first column of of cohort df
retention_matrix_1 = cohort_pivot_type_1.divide(cohort_size_1, axis = 0) #calculate the retention percentage for each cohort

cohort_size_2 = cohort_pivot_type_2.iloc[:,0] # take the first column of of cohort df
retention_matrix_2 = cohort_pivot_type_2.divide(cohort_size_2, axis = 0) #calculate the retention percentage for each cohort
```

```python
#Ideally I would make this a function to reuse but time is money!

with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix_1, 
                mask=retention_matrix_1.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention, Type 1 Transactions', fontsize=16)
    ax[1].set(xlabel='# of months',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohort_size_1).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()

    
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix_2, 
                mask=retention_matrix_2.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention, Type 2 Transactions', fontsize=16)
    ax[1].set(xlabel='# of months',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohort_size_2).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()
```

### Total Revenue per customer

Doing a quick distribution of total revenue per customer in excel showed something close to a normal curve? Is there anything I can learn from it?

Lets split the two groups into those that only did type 1 and type 2 transactions. We want to remove customers that did both as they are not independant samples. What we effectively have now is an A group and and B group!

The two csv files were created in sql using the Other products script

The following resource helped me do the analysis:
https://towardsdatascience.com/statistical-significance-with-the-help-of-python-1fbb318ce216

```python
df_type_1 = pd.read_csv('type_1_tot_rev_per_cust_per_month.csv')
df_type_2 = pd.read_csv('type_2_tot_rev_per_cust_per_month.csv')

frames = [df_type_1,df_type_2]
all_types = pd.concat(frames)

all_types.info()

df_type_2.describe()
```

```python
# How big are these data sets?
print("Type 1 transactions:", len(df_type_1))
print("Type 2 transactions:", len(df_type_2))

#What are the means?
type_1_mean = df_type_1.total_rev.mean()
type_2_mean = df_type_2.total_rev.mean()

print("Type 1 mean:", type_1_mean, len(df_type_1)/(len(df_type_1)+len(df_type_2)) )
print("Type 2 mean:", type_2_mean, len(df_type_2)/(len(df_type_1)+len(df_type_2)) )
```

```python
mean_difference = round(type_2_mean - type_1_mean,2)
print("Test statistic:",mean_difference)
```

For fairness and so we have two fair samples at hand lets take a random sample from each.

**Null Hypothesis:** On average, the total revenue generated by a customers payments is the same for transaction 1 types and transaction 2 types

**Alternative Hypothesis:** On average, the total revenue generated by a customers payments is higher for type 2 transactions

```python
#will this work???

mean_difference = round(type_2_mean - type_1_mean,2)
# a series of all values from the total_sales column
all_revenue_gbp = all_types["total_rev"]

mean_differences = []
# for loop to create our 1000 permutations
for i in range(1000):
    # empty array to represent a random total revenue of type 1 transaction
    group_a = []
    # empty array to represent a random total revenue of type 2 transaction 
    group_b = []
    
    # use a for loop to randomonly place the total sales in one of the two lists
    for total_rev in all_revenue_gbp:
        random_value = np.random.random()
        # randomally assinging each value to a new list
        if random_value > 0.62:
            group_a.append(total_rev)
        elif random_value <= 0.62:
             group_b.append(total_rev)
    mean_group_a = np.mean(group_a)
    mean_group_b = np.mean(group_b)
    # find the itteration mean and append to our mean difference list
    iteration_mean_difference =  mean_group_b - mean_group_a
    mean_differences.append(iteration_mean_difference)

# this is a really good permutation explainer; https://www.jwilber.me/permutationtest/
# use matplotlib to visulize the mean_differences list as a histogram
plt.hist(mean_differences)
plt.show()   
```

```python
total = 0

#for loop to check if random mean diff is greater than our mean difference
for mean_diff in mean_differences:
    if mean_diff >= mean_difference:
        total += 1

#p value calculated by calulating percentage of values greater then the mean difference
p_value = round(total / 1000)

print(p_value)
```

#### Result and Conclusion 
The p_value is less than the decided significance level for this test. It seems we are able to reject the null hypothesis and Type 2 transactions on average do generate more total revenue per customer.

However! Significance is irrelevant without more information (e.g I have no idea of the cost to the business for a type 2 transaction). Also we know it is significant but we dont know why?

This is also the first time I have used python to test significance. I would not report this result to the business until I understood the theory more and code more and had the result peer reviewed by a collegue.

### What about a t-test? 

Will using another method to compute the statistical significance provide the same answer? That we can reject the null hypothesis?

https://towardsdatascience.com/inferential-statistics-series-t-test-using-numpy-2718f8f9bf2f

```python
#Get a random sample of values for each population

a = df_type_1.sample(n=1000)
b = df_type_2.sample(n=1000)
N = 1000
```

```python


## Calculate the Standard Deviation
#Calculate the variance to get the standard deviation

#For unbiased max likelihood estimate we have to divide the var by N-1, and therefore the parameter ddof = 1
var_a = a.var(ddof=1)
var_b = b.var(ddof=1)

#std deviation
s = np.sqrt((var_a + var_b)/2)




## Calculate the t-statistics
t = (a.mean() - b.mean())/(s*np.sqrt(2/N))



## Compare with the critical t-value
#Degrees of freedom
df = 2*N - 2

#p-value after comparison with the t 
p = 1 - stats.t.cdf(t,df=df)


print("t = " + str(t))
print("p = " + str(2*p))
### You can see that after comparing the t statistic with the critical t value (computed internally) we get a good p value of 0.0005 and thus we reject the null hypothesis and thus it proves that the mean of the two distributions are different and statistically significant.


## Cross Checking with the internal scipy function
t2, p2 = stats.ttest_ind(a,b)
print("t = " + str(t2))
print("p = " + str(p2))
```

It seems using the SciPy method also seems to conclude that the type 2 group in statistically significant from the type 1 transactions so we can again reject the Null Hypothesis
